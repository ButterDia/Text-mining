```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from wordcloud import WordCloud
import nltk
from sklearn.metrics import accuracy_score
import string
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
import spacy
from sklearn.linear_model import LogisticRegression
from spacy import displacy
from collections import Counter

nltk.download('punkt')
```

```python
df = pd.read_csv("C:/Users/HP/Downloads/Amazon reviews Iphone15.csv",encoding='utf-8')
df
```

```python
df=df.drop(["id","profileName","text","date"],axis=1)
df
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
translator = str.maketrans("", "", string.punctuation)
df['title'] = df['title'].apply(lambda text: text.translate(translator))
```

```python
df['title'] 
```

```python
df['sentiment_score'] = df['title'].apply(lambda text: sia.polarity_scores(text))
```

```python
df['sentiment_score']
```

```python
df
```

```python
df['neg_score'] = df['sentiment_score'].apply(lambda x: x['neg'])
df['neu_score'] = df['sentiment_score'].apply(lambda x: x['neu'])
df['pos_score'] = df['sentiment_score'].apply(lambda x: x['pos'])
df['compound_score'] = df['sentiment_score'].apply(lambda x: x['compound'])
```

```python
df
```

```python
df['sentiment_label'] = df['sentiment_score'].apply(lambda scores: 'positive' if scores['compound'] > 0 else 'negative' if scores['compound'] < 0 else 'neutral')
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['title'], df['sentiment_label'], test_size=0.3, random_state=42)
```

```python
vectorizer = CountVectorizer(stop_words='english')
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train_vect, y_train)
```

```python
y_train_pred = logreg.predict(X_train_vect)
y_test_pred = logreg.predict(X_test_vect)
```

```python
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
```

```python
print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
```
